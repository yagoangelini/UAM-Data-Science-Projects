## Conte√∫do
O arquivo water_potability.csv cont√©m m√©tricas de qualidade da √°gua para 3.276 corpos d'√°gua diferentes.

1. **Valor de pH:**
O pH √© um par√¢metro importante na avalia√ß√£o do equil√≠brio √°cido-base da √°gua. √â tamb√©m o indicador da condi√ß√£o √°cida ou alcalina do estado da √°gua. A OMS recomendou o limite m√°ximo permitido de pH de 6,5 a 8,5. As faixas de investiga√ß√£o atuais foram de 6,52 a 6,83, que est√£o dentro da faixa dos padr√µes da OMS.

2. **Dureza:**
A dureza √© causada principalmente por sais de c√°lcio e magn√©sio. Esses sais s√£o dissolvidos em dep√≥sitos geol√≥gicos atrav√©s dos quais a √°gua viaja. O per√≠odo de tempo em que a √°gua est√° em contato com o material que produz dureza ajuda a determinar quanta dureza existe na √°gua bruta. A dureza foi originalmente definida como a capacidade da √°gua de precipitar sab√£o causada por C√°lcio e Magn√©sio.

3. **S√≥lidos (Total dissolved solids - TDS):**
A √°gua tem a capacidade de dissolver uma ampla gama de minerais ou sais inorg√¢nicos e alguns org√¢nicos, como pot√°ssio, c√°lcio, s√≥dio, bicarbonatos, cloretos, magn√©sio, sulfatos, etc. Esses minerais produzem sabor indesejado e cor dilu√≠da na apar√™ncia da √°gua. Este √© o par√¢metro importante para o uso da √°gua. A √°gua com alto valor de TDS indica que a √°gua √© altamente mineralizada. O limite desej√°vel para TDS √© de 500 mg/l e o limite m√°ximo √© de 1000 mg/l, prescrito para fins de consumo.

4. **Cloraminas:**
Cloro e cloramina s√£o os principais desinfetantes usados em sistemas p√∫blicos de √°gua. As cloraminas s√£o mais comumente formadas quando a am√¥nia √© adicionada ao cloro para tratar a √°gua pot√°vel. N√≠veis de cloro de at√© 4 miligramas por litro (mg/L ou 4 partes por milh√£o (ppm)) s√£o considerados seguros na √°gua pot√°vel.

5. **Sulfato:**
Os sulfatos s√£o subst√¢ncias naturais encontradas em minerais, solo e rochas. Eles est√£o presentes no ar ambiente, √°guas subterr√¢neas, plantas e alimentos. O principal uso comercial do sulfato √© na ind√∫stria qu√≠mica. A concentra√ß√£o de sulfato na √°gua do mar √© de cerca de 2.700 miligramas por litro (mg/L). Varia de 3 a 30 mg/L na maioria dos suprimentos de √°gua doce, embora concentra√ß√µes muito mais altas (1000 mg/L) sejam encontradas em algumas localiza√ß√µes geogr√°ficas.

6. **Condutividade:**
A √°gua pura n√£o √© um bom condutor de corrente el√©trica, mas sim um bom isolante. O aumento na concentra√ß√£o de √≠ons aumenta a condutividade el√©trica da √°gua. Geralmente, a quantidade de s√≥lidos dissolvidos na √°gua determina a condutividade el√©trica. A condutividade el√©trica (CE) realmente mede o processo i√¥nico de uma solu√ß√£o que permite que ela transmita corrente. De acordo com os padr√µes da OMS, o valor de CE n√£o deve exceder 400 ŒºS/cm.

7. **Carbono Org√¢nico:**
O Carbono Org√¢nico Total (TOC) nas √°guas de nascente vem da mat√©ria org√¢nica natural em decomposi√ß√£o (NOM), bem como de fontes sint√©ticas. TOC √© uma medida da quantidade total de carbono em compostos org√¢nicos em √°gua pura. De acordo com a US EPA, < 2 mg/L como TOC na √°gua tratada/pot√°vel e < 4 mg/Lit na fonte de √°gua que √© usada para tratamento.

8. **Trialometanos:**
THMs s√£o produtos qu√≠micos que podem ser encontrados em √°gua tratada com cloro. A concentra√ß√£o de THMs na √°gua pot√°vel varia de acordo com o n√≠vel de material org√¢nico na √°gua, a quantidade de cloro necess√°ria para tratar a √°gua e a temperatura da √°gua que est√° sendo tratada. N√≠veis de THM de at√© 80 ppm s√£o considerados seguros na √°gua pot√°vel.

9. **Turbidez:**
A turbidez da √°gua depende da quantidade de mat√©ria s√≥lida presente no estado suspenso. √â uma medida das propriedades de emiss√£o de luz da √°gua e o teste √© usado para indicar a qualidade da descarga de res√≠duos em rela√ß√£o √† mat√©ria coloidal. O valor m√©dio de turbidez obtido para Wondo Genet Campus (0,98 NTU) √© inferior ao valor recomendado pela OMS de 5,00 NTU.

10. **Potabilidade:**
Indica se a √°gua √© segura para consumo humano onde 1 significa Pot√°vel e 0 significa N√£o pot√°vel.

#### Random Forest Classifier
Foi desenvolvido pelo professor Leo Breiman em 2001.

√â um algoritmo de aprendizagem supervisionada:
- T√©cnica de machine learning na qual voc√™ treina um modelo usando dados que incluem as caracter√≠sticas e os valores conhecidos para o r√≥tulo, de modo que o modelo aprende a ajustar as combina√ß√µes de recursos para o r√≥tulo. Em seguida, ap√≥s a conclus√£o do treinamento, voc√™ pode usar o modelo treinado para prever r√≥tulos para novos itens para os quais o r√≥tulo √© desconhecido.

A ‚Äúfloresta‚Äù que ele cria √© uma combina√ß√£o (ensemble) de √°rvores de decis√£o, na maioria dos casos treinados com o m√©todo de bagging;
- Usando o m√©todo bagging, cada √°rvore de decis√£o no conjunto √© configurada usando uma amostra com substitui√ß√£o dos dados de treinamento. Estatisticamente falando, √© prov√°vel que a amostra tenha cerca de 64% das inst√¢ncias aparecendo pelo menos uma vez na amostra, essas s√£o chamadas de inst√¢ncias in-bag. O restante das inst√¢ncias, que √© de cerca de 36%, s√£o chamadas de inst√¢ncias fora do saco;
- Durante a configura√ß√£o das √°rvores individuais na Random Forest, um certo grau de randomiza√ß√£o tamb√©m √© aplicado ao selecionar o melhor n√≥ para dividir;
- A sele√ß√£o de dados e features √© aleat√≥ria
- V√°rias √°rvores na floresta

Exemplos de Aplica√ß√£o:
- An√°lise de Fraude de cart√£o de cr√©dito (Fraude/N√£o)
- Diagn√≥stico de Doen√ßa (Doente/Saud√°vel)

#### Classification Report
A __PRECISION__ √© a raz√£o TP / (TP + FP) onde TP √© o n√∫mero de verdadeiros positivos e FP o n√∫mero de falsos positivos. A precision √© intuitivamente a capacidade do classificador n√£o rotular uma amostra negativa como positiva. De todos os valores preditos como positivos, a PRECISION define a propor√ß√£o de valores realmente positivos acertados.

O __RECALL__ √© a raz√£o TP / (TP + FN) onde TP √© o n√∫mero de verdadeiros positivos e FN o n√∫mero de falsos negativos. O RECALL √© intuitivamente a habilidade do classificador em encontrar todas as amostras positivas. De todos os valores realmente positivos, o RECALL define a propor√ß√£o de acertos.

A pontua√ß√£o __F1-SCORE__ pode ser interpretada como uma m√©dia harm√¥nica ponderada da PRECISION e RECALL, onde uma pontua√ß√£o F1-SCORE atinge seu melhor valor em 1 e pior pontua√ß√£o em 0.

O __SUPPORT__ √© o n√∫mero de ocorr√™ncias de cada classe em y_true.

As m√©dias relatadas incluem __MACRO_AVERAGE__ (m√©dia da m√©dia n√£o ponderada por r√≥tulo), __WEIGHTED_AVERAGE__ (m√©dia da m√©dia ponderada de suporte por r√≥tulo) e SAMPLE_AVERAGE (somente para classifica√ß√£o multi-r√≥tulo). A MICRO_AVERAGE (m√©dia do total de verdadeiros positivos, falsos negativos e falsos positivos) s√≥ √© mostrada para multi-r√≥tulo ou multiclasse com um subconjunto de classes, porque de outra forma corresponde √† precis√£o e seria a mesma para todas as m√©tricas
