1 - IDENTIFICAÇÃO DO PROBLEMA DE NEGÓCIO
- Os conjuntos de dados contêm transaçóes feitas por cartões de crédito em setembro de 2013 por titulares de cartões europeus. 
- Este conjunto de dados apresenta as transações que ocorreram em dois dias, onde temos 492 fraudes em 284.807 transaçóes. O 
conjunto de dados é altamente desequilibrado, a classe positiva (fraudes) é responsável por 0,172% de todas as transaçôes. 
- Ele contém apenas variáveis de entrada numéricas que são o resultado de uma transformação PCA. Infelizmente, devido a questões de 
confidencialidade, não podemos fornecer os recursos originais e mais informações básicas sobre os dados. Os recursos VI, V28 
são os componentes principais obtidos com o PCA os únicos recursos que não foram transformados com o PCA são 'Tempo e 
'Quantidade'.
- PCA (Principal Component Analysis = Análise de Componentes Principais) é uma técnica de aprendizagem não supervisionada em que você tem um dataset de alta dimensão (muitas variáveis), e aplica o PCA para diminuir a dimensionalidade. V1, V2, Vn são variáveis/componentes principais gerados a partir da transformação.
- O recurso 'Tempo' contém os segundos decorridos entre cada transação e a primeira transação no conjunto de dados. O recurso 'Amount' 
é o Amount da transação, esse recurso pode ser usado, por exemplo, para estudos diversos. 
- O recurso 'Classe' é a variável de resposta e assume o valor I em caso de fraude e O em caso contrário.

3 - PRÉ-PROCESSAMENTO DOS DADOS
- Esta etapa normalmente leva mais tempo no ciclo, mas para nossos propósitos, a maioria dos conjuntos de dados escolhidos nesta série são limpos.
- Os conjuntos de dados do mundo real são bagunçados e incompletos. As escolhas que fazemos nesta etapa para resolver problemas de dados podem afetar as etapas posteriores e o próprio resultado final (muito cuidado).
- Por exemplo, pode ser complicado lidar com dados ausentes quando não sabemos por que eles estão ausentes(estão faltando por conta de ETL ou de ausência mesmo?).
- É missing mesmo, ou a ausência de Informação tem um significado específico?

```python
#Seria possível nessa etapa fazermos a normalização, padronização dos dados, deixar na mesma escala, binarização (tornar em 0 e 1).
x = df.drop(['Class'], axis=1)
y = df['Class']
X_train, X_val, y_train, y_val = train_test_split(x, y, test_size=0.3, random_state=7)
```

4 - CRIANDO A MÁQUINA PREDITIVA DE DETECÇÃO DE FRAUDE
- Criamos a Máquina Preditiva e ajustamos o modelo aos dados que preparamos para o treinamento. 
- Em seguida, procedemos à classificação com os dados que preparamos para avaliação. 
- Por último, avaliamos o desempenho do modelo com as principais métricas de classificação. 

```python
#Construindo o modelo
#O RandomForestClassifier() possui diversos parâmetros para ajudar no desempenho do modelo
model = RandomForestClassifier(n_estimators=500, max_features=None)

#from xgboost import XGBClassifier
#model = XGBClassifier()
```

```python
#Treinando o modelo
#Demora alguns minutos
model = model.fit(X_train, y_train)
```

```python
#Passando novos dados (dados de teste) para o modelo
y_predict = model.predict(X_val)
```

5 - AVALIANDO A MÁQUINA PREDITIVA

```python
#Comparando gabarito e previsões da máquina
gabarito = pd.DataFrame({'Gabarito': y_val, 'Previsoes_da_maquina':y_predict})
print(gabarito)
```

```python
#Evaluate model
print('Classification metrics: \n', classification_report(y_val, y_predict))

#f1-score é uma métrica média harmônica entre precision/recall. Ela geralmente é mais precisa.

#NORMAL
# Classification metrics: 
#                precision    recall  f1-score   support

#            0       1.00      1.00      1.00     85288
#            1       0.93      0.82      0.87       155

#     accuracy                           1.00     85443
#    macro avg       0.97      0.91      0.94     85443
# weighted avg       1.00      1.00      1.00     85443

#PARÂMETROS ALTERADOS
# Classification metrics: 
#                precision    recall  f1-score   support

#            0       1.00      1.00      1.00     85288
#            1       0.94      0.84      0.88       155

#     accuracy                           1.00     85443
#    macro avg       0.97      0.92      0.94     85443
# weighted avg       1.00      1.00      1.00     85443
```
